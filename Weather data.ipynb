```python
from glob import glob
import os
import pandas as pd
import json
```

```python
w0files = sorted(glob('Data\\weather\\***\\2016-*.json'))
```

```python
w1files = sorted(glob('Data\\weather\\***\\2017-*.json'))
```

```python
airports = ['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
```

```python
weather_data1 = []
for i in w1files:
    with open(i) as f:
        temp = json.load(f)
        for k in temp['data']['weather']:
            tempdata1 = pd.DataFrame(k['hourly'],columns=['windspeedKmph','visibility','weatherCode','precipMM','WindGustKmph','pressure','cloudcover','winddirDegree','humidity','DewPointF','tempF','time','WindChillF'])
            tempdata1['date'] = k['date']
            for a in airports:
                if a in i.split('\\'):
                    tempdata1['place'] = a 
            weather_data1 = pd.DataFrame(weather_data1.append(tempdata1))
weather_data1.reset_index()
```

```python
weather_data1.to_csv('Data\\weather17.csv')
```

```python
weather_data0 = []
for i in w0files[1:]:
    with open(i) as f:
        temp = json.load(f)
        for k in temp['data']['weather']:
            tempdata0 = pd.DataFrame(k['hourly'],columns=['windspeedKmph','visibility','weatherCode','precipMM','WindGustKmph','pressure','cloudcover','winddirDegree','humidity','DewPointF','tempF','time','WindChillF'])
            tempdata0['date'] = k['date']
            for a in airports:
                if a in i.split('\\'):
                    tempdata0['place'] = a 
            weather_data0 = pd.DataFrame(weather_data0.append(tempdata0))
weather_data0.reset_index()
```

```python
weather_data0.to_csv('Data\\weather16.csv')
```

```python
weather_df = pd.concat([weather_data0,weather_data1])
weather_df.to_csv('Data\\Weather_Data.csv')
```
